# Hierarchial Clustering

### Single linkage

```python
clustering1 = Hierarchial(distance_matrix=distance_matrix, linkage='single', n_clusters=n_clusters)
clustering1.clustering()
```

```python
fig = plt.figure(figsize=(10, 10))
colors = ['red', 'blue', 'green', 'yellow', 'grey', 'black', 'pink', 'orange', 'purple', 'brown']
for i in range(len(clustering1.cluster_list)):
    X = [data[keys[0]][point] for point in clustering1.cluster_list[i].points]
    Y = [data[keys[1]][point] for point in clustering1.cluster_list[i].points]
    plt.scatter(X, Y, c=colors[i])
```

### Complete linkage

```python
clustering2 = Hierarchial(distance_matrix=distance_matrix, linkage='complete', n_clusters=n_clusters)
clustering2.clustering()
```

```python
fig = plt.figure(figsize=(10, 10))
colors = ['red', 'blue', 'green', 'yellow', 'grey',
          'black', 'pink', 'orange', 'purple', 'brown']
for i in range(len(clustering2.cluster_list)):
    X = [data[keys[0]][point] for point in clustering2.cluster_list[i].points]
    Y = [data[keys[1]][point] for point in clustering2.cluster_list[i].points]
    plt.scatter(X, Y, c=colors[i])
```

### Average linkage

```python
clustering3 = Hierarchial(distance_matrix=distance_matrix, linkage='average', n_clusters=n_clusters)
clustering3.clustering()
```

# DBSCAN Clustering

## Eps = 0.2, minPoints = 6

```python
clustering4 = DBSCAN(distance_matrix=distance_matrix, eps=0.2, MinPoints=6)
clustering4.clustering()
```

```python
fig = plt.figure(figsize=(10, 10))
colors = ['red', 'blue', 'green', 'yellow', 'grey', 'pink', 'orange', 'purple', 'brown']
print(len(clustering4.cluster_list))
for i in range(len(clustering4.cluster_list)):
    X = [data[keys[0]][point] for point in clustering4.cluster_list[i].points]
    Y = [data[keys[1]][point] for point in clustering4.cluster_list[i].points]
    plt.scatter(X, Y, c=colors[i])

X = [data[keys[0]][point] for point in clustering4.outliers]
Y = [data[keys[1]][point] for point in clustering4.outliers]
plt.scatter(X, Y, c='black')
```
